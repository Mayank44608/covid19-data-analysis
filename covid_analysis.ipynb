```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../dataset/covid_19_clean_complete.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
country_cases = df.groupby("Country/Region")["Confirmed"].sum().sort_values(ascending=False)

country_cases.head(10)
```

```python
country_cases.head(10).plot(kind="bar")

plt.title("Top 10 Countries with COVID Cases")
plt.show()
```

```python
date_cases = df.groupby("Date")["Confirmed"].sum()

date_cases.plot()

plt.title("COVID Cases Over Time")
plt.show()
```
